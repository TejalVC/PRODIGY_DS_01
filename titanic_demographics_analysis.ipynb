Create a bar chart or histogram to visualise the distribution of a categorical and continuous variable , such as the distribution of ages or genders in a population.

Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
titanic_data=pd.read_csv("/content/Titanic.csv")
```

```python
# Data Exploration
print(titanic_data.head())  # Check the first few rows of the dataset
```

```python
print(titanic_data.info())  # Get information about the dataset including missing values
```

```python
# Data Cleaning
titanic_data['age'].fillna(titanic_data['age'].median(),inplace=True)
```

```python
# Set up the style for visualizations
sns.set_style("whitegrid")
sns.set_palette("pastel")
```

Visualization

```python
#Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['age'], bins=20, kde=True, color='skyblue', linewidth=1.5)
plt.title('Distribution of Ages among Titanic Passengers', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Distribution of Gender
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_data, x='sex')
plt.title('Distribution of Gender among Titanic Passengers', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
